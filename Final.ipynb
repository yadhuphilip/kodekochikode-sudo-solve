## Importing the required modules

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
```

```python
xls1 = pd.read_excel("C:/Users/MyPc/Desktop/hifx/data/2013_manhattan.xls")
xls2 = pd.read_excel("C:/Users/MyPc/Desktop/hifx/data/2014_manhattan.xls")
xls3 = pd.read_excel("C:/Users/MyPc/Desktop/hifx/data/2015_manhattan.xls")
xls4 = pd.read_excel("C:/Users/MyPc/Desktop/hifx/data/2016_manhattan.xls")
xls5 = pd.read_excel("C:/Users/MyPc/Desktop/hifx/data/2017_manhattan.xls")

df1 = pd.DataFrame(xls1)
df2 = pd.DataFrame(xls2)
df3 = pd.DataFrame(xls3)
df4 = pd.DataFrame(xls4)
df5 = pd.DataFrame(xls5)
```

### Understanding the data

```python
df = [df1,df2,df3,df4,df5]
df1.info()
```

```python
df5.describe(include='all').T
```

### working on data

```python
for ifg in df:
    
    ls1 = ifg['NEIGHBORHOOD\n'].unique()
    lsdict={}
    a=0
    for i in ls1:
        lsdict[i]=a+1
        a=a+1
    lsdict 

    ifg['NEIGHBORHOOD\n'] = ifg['NEIGHBORHOOD\n'].map(lsdict)
    ifg['BUILDING CLASS CATEGORY\n']=ifg['BUILDING CLASS CATEGORY\n'].str[:2]
    ifg['BUILDING CLASS CATEGORY\n']=ifg['BUILDING CLASS CATEGORY\n'].astype(int)
    ifg['NEIGHBORHOOD\n'].value_counts()
```

```python
dfi =[df1,df2,df3,df4]
for i in dfi:
    ls=i['BUILDING CLASS AT PRESENT\n'].unique()
    lsd={}
    a=0
    for ai in ls:
        lsd[ai]=a+1
        a=a+1
    i['BUILDING CLASS AT PRESENT\n']=i['BUILDING CLASS AT PRESENT\n'].map(lsd)
```

```python
df5['BUILDING CLASS AT PRESENT\n']=df5['BUILDING CLASS AS OF FINAL ROLL 17/18']
ls=df5['BUILDING CLASS AT PRESENT\n'].unique()
lsd={}
a=0
for ai in ls:
    lsd[ai]=a+1
    a=a+1
df5['BUILDING CLASS AT PRESENT\n']=df5['BUILDING CLASS AT PRESENT\n'].map(lsd)
```

```python
df_train=df1.append(df2)
df_train=df_train.append(df3)
df_train=df_train.append(df4)
df_train=df_train.append(df5)
```

```python
feature_cols = ['NEIGHBORHOOD\n','BUILDING CLASS CATEGORY\n','BUILDING CLASS AT PRESENT\n','LAND SQUARE FEET\n','BLOCK\n','LOT\n','RESIDENTIAL UNITS\n','COMMERCIAL UNITS\n','TOTAL UNITS\n','LAND SQUARE FEET\n','GROSS SQUARE FEET\n','YEAR BUILT\n']
x=df_train[feature_cols]
y=df_train['SALE PRICE\n']
```

### Splitting the train and test datas

```python
xTrain, xTest, yTrain, yTest = train_test_split(x, y, train_size = 0.87, random_state = 0)
```

### Fitting the model

```python
lr=LinearRegression()
lr.fit(xTrain,yTrain)
rr=Ridge(alpha=0.1)
rr.fit(xTrain,yTrain)
```

```python
yPredict=lr.predict(xTest)
yp=rr.predict(xTest)
```

```python
print(lr.score(xTest,yTest))
print(rr.score(xTest,yTest))
```
